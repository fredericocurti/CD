<font size=5>**Introdução**</font><p>
Esse projeto consiste na realização de um teste AB, uma importante ferramenta usada por serviços no mundo todo. O Teste AB consiste da realização implícita de um experimento no qual duas (ou mais) vers~oes de um produto com alterações entre si são distribuídas aleatoriamente à um grupo de usuários, e através de ferramentas de medição (nesse caso, JavaScript) e armazenamento de dados (Firebase) é possível inferir uma análise estatística sobre qual versão apresenta maior eficiência. Nesse trabalho, fizemos proveito do nosso já disponível portfólio de projetos, resultado do primeiro projeto da disciplina Co-design de Aplicativos, e promovemos algumas alterações entre eles. Especificamente, alteramos as abas que comportam os projetos, como ilustram as imagens abaixo: <br><br>

Essa série de dados registrou, para cada visitante do site, a versão ( A ou B ) e o projeto a que eles foram submetidos e o projeto, ambos definidos aleatoriamente através da função random do JavaScript. Também foi registrado o tempo no início e no fim da missão, no qual a variação do tempo, ou seja, $\Delta T$, será a nossa variável aleatória de interesse durante a análise.

____

A seguir estão todas as etapas do processo de filtração e rearranjo dos dados, o que inclui a subtração dos tempos final e inicial para o delta T:




```python
import pandas as pd
import itertools
import matplotlib.pyplot as plt
from scipy import stats
from numpy import *

%matplotlib inline
```

```python
data = pd.read_json("dadosup.json")
data = data.transpose()
data = data.reset_index()
data = data.drop("index",axis=1)
agrupado = data.groupby(by=["user_id", "versao", "projeto"])
users = set(data.user_id)
users = list(users)
projetos = set(data.user_id)
projetos = list(projetos)
resultados = []
for u in users:
    for versao in ["A", "B"]:
        for proj in ["amplificador","jogo da velha"]:
            if (u, versao, proj) in agrupado.groups.keys():
                grupo1 =  agrupado.get_group((u, versao, proj))
                grupo1 = grupo1.sort_values(by="tempo")

                if len(grupo1[grupo1.tipo=="final"])>= 1 and len(grupo1[grupo1.tipo=="inicio"]) >= 1:
                    delta_t = grupo1[grupo1.tipo=="final"].iloc[-1].tempo - grupo1[grupo1.tipo=="inicio"].iloc[0].tempo
                    resultados.append({"user_id":u, "delta_t":delta_t, "versao":versao, "projeto":proj})

grupo1.sort_values(by="tempo")
data_resumo = data.from_records(resultados)
data_resumo.delta_t/=1000
dados = data_resumo
dados
```

___
A tabela acima é o resultado final que será explorado a partir de agora. Ela já filtrou entradas inválidas como repetições e tempos incompletos, embora um mesmo usuário possa realizar o teste mais de uma vez para versões diferentes. Infelizmente nesse processo **há uma perda considerável de dados,** e o que antes parecia um banco de dados enorme se reduziu à outro muito menor, o que influencia diretamente nas conclusões que serão tomadas. Uma prova disso está na imagem abaixo, que mostra que nosso link encurtado teve, até o presente, **205 acessos**, porém apenas 44 entradas válidas foram registradas, algo muito distante do valor esperado. <img src="5.png">

Abaixo é realizado outro filtro de dados aberrantes, nesse caso para ocorrências nas quais o tempo percorrido no teste foi maior do que 200 segundos. Estes estão demonstrados no histograma abaixo, segregados pelas versões. É interessante lembrar que os testes foram conduzidos em usuários de diferentes contextos e possívelmente de diferentes práticas com tecnologia, o que pode ter contribuído para a aparição de alguns dos dados mais extremos, possívelmente filtrados.
<br>
Curiosamente, a ideia era que o histograma aparentasse uma distribuição normal, porém, como o número de amostras é muito baixo, isso não ocorreu. 

```python
tempos_A = list(data_resumo[(data_resumo.versao=="A") & (data_resumo.delta_t < 200)].delta_t)
tempos_B = list(data_resumo[(data_resumo.versao=="B") & (data_resumo.delta_t < 200)].delta_t)

# plt.hist(tempos_A, bins=15,label = "Distribuição de delta T na versão A")
# plt.hist(tempos_B, bins=15, alpha=0.5, label = "Distribuição de delta T na versão B")
# plt.legend(loc="best")
# plt.show()

x = arange(0,60,0.1)
mA = mean(tempos_A)
nA = len(tempos_A)
mB = mean(tempos_B)
nB = len(tempos_B)
nAB = nA + nB
varA = 0
for i in tempos_A:
    varA += ((i - mA)**2)/(nA -1)  
stdA = (varA)**0.5
varB = 0
for i in tempos_B:
    varB += ((i - mB)**2)/(nB -1) 
stdB = (varB)**0.5
trefB = stats.t.pdf(x,nB,mB,stdB)
trefA = stats.t.pdf(x,nA,mA,stdA)

fig = plt.figure(figsize=(15,8))
plt.plot(x,trefB, c = "gold",label = "Distribuição T de B para referência",lw=3)
plt.plot(x,trefA, c = "salmon",label = "Distribuição T de A para referência",lw=3)
plt.hist(tempos_A, bins=15,normed=True,color="teal",stacked=True,label = "Versão A")
plt.hist(tempos_B, bins=15,normed=True,color="springgreen",stacked=True,alpha=0.5, label = "Versão B")
plt.legend(loc='best')
plt.grid(True)
```

As distribuições T de B e A acima são fruto da aproximação para os valores de σ com S da população através da seguinte fórmula:
<img src="S.png">

Abaixo está explícita a distribuição de nossas amostras no quesito versão e projeto. A escolha aleatória do projeto, embora não traga grande valor para a análise estatística, aumenta o dinamismo dos testes, visto que fica mais difícil "roterizar" o processo. Outra curiosidade é que **embora o JavaScript estivesse distribuindo as versões e os projetos com a mesma chance para todos os usuários, houve uma visível discrepância entre a distribuição de ambas as variáveis**. Isso provavelmente se dá pelo baixo número amostral e pelos erros humanos, que também são aleatórios.

```python
ct = pd.crosstab(dados.projeto,dados.versao)
ct.T.plot.bar(stacked=True,colormap = "summer",title="Distribuição amostral entre projeto e versão (A/B)",grid=True)
```

Agora será necessário definir as hipóteses que serão testadas no nosso caso, como não temos um valor suficiente de amostras, optamos pela realização de um **Teste T de Student** para rejeitar ou não a hipótese nula, definida à seguir:
<img src="H0-H12.png">
Ou seja, **a hipótese nula (Ho)** é de que mudar a versão de A para B não traz diferença alguma ou até aumenta o tempo médio levado para encontrar o projeto, enquanto **a hipótese alternativa (H1)** diz justamente o contrário: mudar a versão do site de A para B traria uma redução no tempo médio de procura pelo projeto.
<br><br>
Como estamos lidando com duas médias diferentes, com tamanhos amostrais diferentes, foi necessário o uso da seguinte fórmula para adequação das distribuições ao teste t para a diferença de duas médias.
<img src="Sp.png" width="400px">
No qual Sp² é a variança da diferença amostral. Com isso, é possível calcularmos o **valor t** através da seguinte fórmula, para uma significância de 10%
<img src="alpha.png" width="100px">
<img src="t0.png" width="200px">

```python
varAB = ((nA - 1)*stdA**2 + (nB - 1)*stdB**2)/(nA+nB-2)
stdAB = varAB**0.5
normx = arange(-40,80,0.01)
t0 = (mB - mA)/(stdAB*((1/nA) + (1/nB))**0.5)
print("To =",t0)
```

Agora, sabendo a significância $\alpha$, podemos encontrar o valor T crítico, ou seja, a partir do qual rejeitamos a hipótese nula. Nesse caso, de acordo com a tabela, para $\alpha = 0.1$, temos:
<img src="tt.png">

Com isso, temos que:

<img src="naorejeita.png",width="500px">
<img src="senao.png",width="350px">
<br><br>
O gráfico abaixo mostra como isso se comporta numa distribuição t padrão, na qual podemos ver que o valor que encontramos para o tempo médio entre as duas versões está dentro do intervalo aceitável.

```python
alpha = 0.1
tnorm = 1.306
normtx = arange(-3, 3, 0.01)
normtt = stats.t.pdf(normtx, nAB - 2, loc=0, scale=1)
fig, (ax1) = plt.subplots(1, 1, sharex=True, figsize=(16,8))
#ax1.fill_between(normtx, 0, normtt, where=(normtx < -tnorm) ,color="grey",alpha = 0.5)
ax1.fill_between(normtx, 0, normtt, where=(normtx > tnorm) ,color="grey",alpha = 0.5)
plt.axvline(t0, lw=3, color="purple", label='valor T das médias AB',ymax = normtt[310]*2.35)
plt.axvline(tnorm, lw=2, color="red", label='+Tc', ls='dashed')
#plt.axvline(-tnorm, lw=2, color="red", label='+Tc', ls='dashed')
plt.plot(normtx, normtt, c="springgreen",lw="3",label ="Curva T padrão")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

___
<font size=5>**Conclusão**</font><p>

Visto tudo, com o **valor μ(AB) dentro do intervalo de confiança,** não rejeitamos a hipótese nula. Ou seja, a mudança de versão não permite uma conclusão significativa sobre uma melhora no tempo médio. Com isso, imaginamos que mesmo se tivessemos obtido um número maior de amostras, o resultado não seria tão diferente. Logo, ter as abas colapsadas ou expandidas trouxe pouco ou nenhum benefício na hora de encontrar determinado projeto.
