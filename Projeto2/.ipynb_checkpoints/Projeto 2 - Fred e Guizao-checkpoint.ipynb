<font size=5>**Introdução**</font><p>
Esse projeto consiste na realização de um teste AB, uma importante ferramenta usada por serviços no mundo todo. O Teste AB consiste da realização implícita de um experimento no qual duas (ou mais versões) de um produto com alterações entre si são distribuídas aleatoriamente à um grupo de usuários, e através de ferramentas de medição (nesse caso, JavaScript) e armazenamento de dados (Firebase) é possível inferir uma análise estatística sobre qual versão apresenta maior eficiência. Nesse trabalho, fizemos proveito do nosso já disponível portfólio de projetos, resultado do primeiro projeto da disciplina Co-design de Aplicativos, e promovemos algumas alterações entre eles. Especificamente, alteramos as abas que comportam os projetos, como ilustram as imagens abaixo: <br><br>

E uma boa métrica para medir a eficácia de cada versão é o <font color ="red"> tempo. </font>

Também foi registrado o tempo no início e no fim da missão, no qual a variação do tempo, ou seja, $\Delta T$, será a nossa variável aleatória de interesse durante a análise.

____

A seguir estão todas as etapas do processo de filtração e rearranjo dos dados, o que inclui a subtração dos tempos final e inicial para o delta T




```python
import pandas as pd
import itertools
import matplotlib.pyplot as plt
from scipy import stats
from numpy import *

%matplotlib inline
```

```python
data = pd.read_json("dados.json")
data = data.transpose()
data = data.reset_index()
data = data.drop("index",axis=1)
agrupado = data.groupby(by=["user_id", "versao", "projeto"])
users = set(data.user_id)
users = list(users)
projetos = set(data.user_id)
projetos = list(projetos)
resultados = []
for u in users:
    for versao in ["A", "B"]:
        for proj in ["amplificador","jogo da velha"]:
            if (u, versao, proj) in agrupado.groups.keys():
                grupo1 =  agrupado.get_group((u, versao, proj))
                grupo1 = grupo1.sort_values(by="tempo")

                if len(grupo1[grupo1.tipo=="final"])>= 1 and len(grupo1[grupo1.tipo=="inicio"]) >= 1:
                    delta_t = grupo1[grupo1.tipo=="final"].iloc[-1].tempo - grupo1[grupo1.tipo=="inicio"].iloc[0].tempo
                    resultados.append({"user_id":u, "delta_t":delta_t, "versao":versao, "projeto":proj})

grupo1.sort_values(by="tempo")
data_resumo = data.from_records(resultados)
data_resumo.delta_t/=1000
dados = data_resumo
dados
```

A tabela acima é o resultado final que será explorado a partir de agora. Ela já filtrou entradas inválidas como repetições e tempos incompletos, embora um mesmo usuário possa realizar o teste mais de uma vez para versões diferentes. Infelizmente nesse processo há uma perda considerável de dados, e o que antes parecia um banco de dados enorme se reduziu à outro muito menor, o que influencia diretamente nas conclusões que serão tomadas

```python
tempos_A = list(data_resumo[data_resumo.versao=="A"][data_resumo.delta_t < 200].delta_t)
tempos_B = list(data_resumo[data_resumo.versao=="B"][data_resumo.delta_t < 200].delta_t)

plt.hist(tempos_A, bins=15)
plt.hist(tempos_B, bins=15, alpha=0.5)
plt.show()
```

```python
ct = pd.crosstab(dados.projeto,dados.versao)
ct.T.plot.bar(stacked=True,colormap = "summer",title="Distribuição amostral entre projeto e versão (A/B)",grid=True)
```

```python
x = range(0,50)
fitB = stats.t.fit(dados.delta_t[dados.versao == "B"].astype(float))
tref = stats.t.pdf(x,fitB[0],fitB[1],fitB[2])
trefA = stats.t.pdf(x,fitA[0],fitA[1],fitA[2])
fig = plt.figure(figsize=(15,8))
plt.plot(x,tref, c = "gold",label = "Distribuição T de B para referência",lw=3)
plt.hist(tempos_A, bins=15,normed=True,color="teal",stacked=True,label = "Versão A")
plt.hist(tempos_B, bins=15,normed=True,color="springgreen",stacked=True,alpha=0.5, label = "Versão B")
plt.legend(loc='best')
plt.grid(True)
```

```python
mA = mean(tempos_A)
nA = len(tempos_A)
mB = mean(tempos_B)
nB = len(tempos_B)
nAB = nA + nB

varA = 0
for i in tempos_A:
    varA += ((i - mA)**2)/(nA -1)
    
stdA = (varA)**0.5

varB = 0
for i in tempos_B:
    varB += ((i - mB)**2)/(nB -1)
    
stdB = (varB)**0.5

varAB = ((nA - 1)*stdA**2 + (nB - 1)*stdB**2)/(nA+nB-2)

stdAB = varAB**0.5

normx = arange(-40,80,0.01)
```

```python
t0 = (mB - mA)/(stdAB*((1/nA) + (1/nB))**0.5)
```

```python
normABAB = stats.t.pdf(normx, nAB - 2, loc=t0, scale=stdAB)

fig = plt.figure(figsize=(15,8))

plt.axvline(t0, lw=2, color="purple", label='Média da amostra', ls='dashed')

plt.plot(normx, normABAB, c="purple")
plt.show()
```
